```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X = np.load('features_AC.npy')
y = np.load('y_response.npy')
```

```python
# plot tsne
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
```

```python
X_2d = tsne.fit_transform(X)
```

```python
# Plot the feature importances of the forest
plt.figure(figsize=(40,30))

B_SIZE = 30
axis_font = {'fontname':'Arial', 'size':'30'}
plt.rc('font', size=B_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=B_SIZE+10)     # fontsize of the axes title
plt.rc('axes', labelsize=B_SIZE+10)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=B_SIZE-5)    # fontsize of the tick labels
plt.rc('ytick', labelsize=B_SIZE-5)    # fontsize of the tick labels
plt.rc('legend', fontsize=B_SIZE)    # legend fontsize
plt.rc('figure', titlesize=B_SIZE)  # fontsize of the figure title
#plt.rc('xlabel', titlesize=B_SIZE+10)
#plt.rc('ylabel', titlesize=B_SIZE+10)
labels = [1, 2, 3]
class_label = ['Left Hand', 'Right Hand', 'Feet']

plt.figure(figsize=(40,30))
plt.title('t-SNE for hybrid dataset', fontweight='bold')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
colors = 'r', 'g', 'b'
for i, c, label in zip(labels, colors, class_label):
    plt.scatter(X_2d[y == i, 0], X_2d[y == i, 1], c=c,
                label=label, alpha=0.8, linewidths=2, edgecolors ='k',
               s=1000)
plt.legend()
plt.savefig('tsne_hyb_ds.png', bbox_inches='tight')
plt.show()
```

```python
X_tp = np.transpose(X)
corr = np.corrcoef(X_tp)
```

```python
corr.shape
```

```python
plt.figure(figsize=(40,30))
B_SIZE = 30
axis_font = {'fontname':'Arial', 'size':'30'}
plt.rc('font', size=B_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=B_SIZE+10)     # fontsize of the axes title
plt.rc('axes', labelsize=B_SIZE+10)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=B_SIZE-5)    # fontsize of the tick labels
plt.rc('ytick', labelsize=B_SIZE-5)    # fontsize of the tick labels
plt.rc('legend', fontsize=B_SIZE)    # legend fontsize
plt.rc('figure', titlesize=B_SIZE)  # fontsize of the figure title
plt.title('Correlation Heatmap', fontweight='bold')
plt.xlabel('Features')
plt.ylabel('Features')
ret_y  =  plt.imshow(corr, cmap='hot', interpolation='nearest')
plt.colorbar(ret_y)
plt.savefig('pearson_corr_mat_after_feat_select.png', bbox_inches='tight')
plt.show()
```
